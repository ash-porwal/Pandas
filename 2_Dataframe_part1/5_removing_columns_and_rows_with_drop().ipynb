```python
import pandas as pd
```

```python
df = pd.read_csv('imdb_top_1000.csv')
df.head(3)
```

```python
# to read specific columns we can even pass it in dataframe
df[['Genre', 'Runtime']]
```

```python
# Suppose I want to remove some columns from my original dataframe
# then drop method is going to help in that
# need to specify the number of columns we want to remove
df.drop(columns=['Genre', 'Runtime'])
```

```python
# we can also drop labels(index or rows)
# then we want to pass aindex number
df.drop(labels=[1,2,3,4])
```

```python
# we also dont need to specify it is for labels
# because axis is 0 by default
df.drop([0,5])
```

```python
# if we want to remove the columns with axis
# then we need to provide column name and axis set to 1.
df.drop(['Series_Title'], axis=1)
```
