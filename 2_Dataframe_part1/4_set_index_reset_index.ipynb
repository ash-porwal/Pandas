### Let's learn how we can modify index

```python
import pandas as pd
```

```python
cols = ['Series_Title', 'IMDB_Rating', 'No_of_Votes']
```

```python
df = pd.read_csv("imdb_top_1000.csv", usecols=cols)
df
```

### set_index() method
it takes two arguments
* `drop` - removes the column which has now become the index (True by default).
* `inplace` - going to update the dataframe (False by default)

```python
# settting Series_Title column as index
df.set_index('Series_Title', drop=False, inplace=False)
# so wierd thing is we can now see two columns with Series_Title name.
# so setting drop to True will remove that duplicated column.
```

```python
df.set_index('Series_Title') # by default drop is True
```

```python
# we can even change the Title of the index
df.index.name = 'Custom title'
df
```

### reset_index()
* Resets the indec

```python
df.reset_index()
```
