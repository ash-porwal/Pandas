`````python
import pandas as pd
`````

`````python
df = pd.read_csv('Pokemon.csv', index_col='#')
df.head()
`````

### replace()

* So, in map method we can replace a certain key with a certain value by passing in a dictionary.
* but we couldn't pass the values correctly we will have a lot of missing values.
* so, using `replace` we can replace a single value in our dataframe.

`````python
df['Legendary'].replace({True: 'Yes'}, inplace=True)
df['Legendary'].value_counts()

# Notice in our new DataFrame we don't have Boolean - True, instead we have Yes.
`````

`````python
# if we want to replace values from entire dataframe regardless of columns then we can do like this
# we need to pass the dictionary to the new value on whole dataframe
df.replace({'Grass': '~~~~~~', 'First': '````'}, inplace=True) # if we specify the value which doesn't exists then nothing is going to happen.
df.head()
`````
