```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df.info()
```

```python
# I want to create a new column and that column should contains the length of name of each pokemon
df['Name_len'] = df['Name'].apply(len)
df.head()
# so, apply function takes a method name as its parameter, 
# and that method name will be  applied to all the rows of that selected column.
```

```python
# but we can create our own function and pass it to apply method.

# lets delete the column which we created 
del df['Name_len']
df.head()
```

```python
# defining custom function
def strength(attack: int)->str:
    if attack>50:
        return "Strong"
    elif attack > 100:
        return "Average"
    else:
        return "Weak"
```

```python
# Passing our function in apply
df['Strength'] = df['Attack'].apply(strength)
df.head()
```

```python
# apply can also be used with multiple columns
def score(row):
    attack: int = row['Attack']
    defense: int = row['Defense']
    speed: int = row['Speed']
    
    return sum([attack, defense, speed])
# so if we want it to apply on multiple columns we need to set the axis to 1.
df['Score'] = df.apply(score, axis=1)
"""
When we call score function and noticed it expects one argument which is - row
so, we are calling this function like: df.apply(score, axis=1)

in here -   score function will receive either a Series representing each column 
            or a row of DataFrame, it will depends upon axis parameter.
            The default behavior is axis=0, meaning the function will be applied to each column.

            So, when axis=0, it gets each column as a Series
            when axis=1, it gets each row from all columns

"""
df.head()
```
