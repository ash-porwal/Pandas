```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv', )
df.head()
```

### where()
as we already know with conditional statements we can filter out things from our dataframe.
Similar things we can do with where() - but it just returns `NaN` if condition is False.

```python
# lets get the generation 2 pokemon using where
df.where(df['Generation'] == 2) # we can use dropna() to see what we filtered
```

```python
# so, lets try to use dropna
df.where(df['Generation'] == 2).dropna() # now all the missing values are dropped and we just got pokemon with Generation 2.
```

```python
# and also instead of fill with NaN if it doesn't met the condition
# we can also fill it with our custom fields, `other` is the parameter
df.where(df['Generation'] == 2, other='-') # here I've filled it with dash(-) instead of NaN.
```
