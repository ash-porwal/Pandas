```python
import pandas as pd
```

In this we are going to know what Hierarchical indexing is - (also known as multi-indexing)

```python
df = pd.read_csv("Pokemon.csv")
df
```

In Pandas we can create multi-index by passing in a list of columns names into a `set_index()`

```python
# Lets set Name and Type 1 column as index
df.set_index(["Name", "Type 1"], inplace=True)
df
```

```python
# lets get the index
df.index
# Oh! we got multi index - in a tuple - means first element we can access First index name 
# with 0 and second with 1 and so on..
```

lets get the Name index back - with `get_level_values()` -  this helps us to specify which index you want to get back

```python
df.index.get_level_values(1) # This way we get the Name index list
```

We can also swap the levels using `swaplevel()`

```python
df = df.swaplevel()
df # Now, Type 1 is first value of index and Name column is 2nd value.
```
