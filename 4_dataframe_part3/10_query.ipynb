```python
import pandas as pd
```

```python
new_df = pd.read_csv("Pokemon.csv")
new_df.head(5)
```

Now, I will try to get unique Type2 column value of those Pokemons.
I can do this with unique function.
`Syntax` - df["col_name"].unique()

```python
# Getting unique Type 2
new_df["Type 2"].unique()
```

Now, If I want to find some particular value, like -> Where Pokemon got 100 Attack, So in that case we can use `query` function

```python
# Lets check how we can use this query function
new_df.query?
```

```python
new_df.query('Attack == 100')
```

```python
# we can use operators with query as well like -
new_df.query('Attack == 100 and HP == 100' )
```
