```python
import pandas as pd
```

```python
df = pd.read_csv("people2.csv", sep=";")
df
```

```python
df.info()
```

Convert Salary Column into integer

Issues when converting this column into integer is -
* there is comma in the value
* and a symbol(€) in value

```python
formatted = df['Salary'].str.replace('€', '').str.replace(',', '').astype(int) / 100
df['Salary'] = formatted
df
```

`pivot_table` is same as pivot but we can perform additional operations in it.

Get the sum of the salary for each person

```python
df.pivot_table(
    values='Salary', # Which value you want to use
    columns='Name', # Which columns to perform the aggfunc on
    aggfunc='sum' # which opertaion you want to use (mean operation is default)
)
```

Get the average salary for each day for all the tables

```python
df.pivot_table(
    values='Salary',
    columns='Day'
).round(2)
```

Find what the average salary is per day per gender

```python
df.pivot_table(
    index='Day',
    values='Salary',
    columns = 'Gender',
    aggfunc='mean'
)
```
