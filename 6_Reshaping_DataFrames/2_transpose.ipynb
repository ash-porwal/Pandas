```python
import pandas as pd
```

```python
df = pd.read_csv("exams.csv")
```

```python
df.head()
```

`transpose()` -> Invert the columns and Indices - means all column will become index

```python
df.transpose().head()
```

```python
# Another and shorter way of doing trnanspose is - referring it as T
df.T.head()
```

`Warning` - When we use transpose(), our data might get change

```python
df.info() # This has 5 objects and 3 integer - so the data type got changed
```

```python
transposed_df = df.T
transposed_df.info() # after transposing - all data is now an object and no integers are there - 
# when column contains mix value of string and integer, so that column becomes object
```

```python
# Now what if we transponse back the transposed dataframe
t2_df = transposed_df.T
t2_df.head() # we can see everything got back to normal as the original data.
```

```python
# but if we run info on t2_df
t2_df.info() # now everything is an object
```
