The melt() function in Pandas is used to transform or "unpivot" a DataFrame from a wide format to a long format.

To do this we will use
`melt()` - which used to unpivot a DataFrame from wide to long format
* `id_vars` - are what we want to preserve, that we don't want to use as values
* `values_vars` - are what we want to melt into variables

```python
medals_df.melt(
    id_vars='Country',
    value_vars=['2021','2022','2023', '2024']
)
# if we notice then we have now two new columns, variables and value
```

```python
medals_df # melt doesn't perform inplace operation
```

```python
medals_df.melt(
    id_vars='Country',
    var_name='Year', # changing the variable name
    value_vars=['2021','2022','2023', '2024']
)
```
