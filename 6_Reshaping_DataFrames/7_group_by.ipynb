With `groupby()` we can reshape our dataset

```python
pokemons = pd.read_csv('/home/ashporwal/hdd/lab/programmes/Python/Pandas/4_dataframe_part3/Pokemon.csv', index_col='#')
pokemons
```

Lets take a look at the values with have in `Type 1`

```python
pokemons['Type 1'].value_counts() # These are the values we want to group by
```

in above output we can see we have 112 Water type pokemon, 98 Normal and so on ...

So, Now I will try to group these values with `groupby()`

```python
poke_groups = pokemons.groupby(by='Type 1')

poke_groups # doesn't returns anything just an object
```

```python
# we can check all the values of above poke_groups variable
poke_groups.groups # This will return group of each Type with the indices as value
```

Bug is the name of the pokemon, but this is not correct way to read the dataframe

```python
poke_groups.get_group('Grass').head() # so, we can get the group by referring the type
```

groupby with more than 1 column

```python
poke_groups = pokemons.groupby(by=['Type 1', 'Type 2'], dropna=False)
poke_groups.groups
```

Once we have grouped any column, then it is pretty easy to perform operations on that dataframe.

```python
max_poke = poke_groups.max().round(2)
max_poke
```

Now, how we can use `agg()` with group

```python
# we are grouping by Type 1 and performing agg, and in agg we can pass the column
# name and the opertaion we want to perform
pokemons.groupby('Type 1').agg(
    {
        'Total': ['max', 'min', 'mean'] # name and the opertaion we want to perform

    }
).round(2)
```
