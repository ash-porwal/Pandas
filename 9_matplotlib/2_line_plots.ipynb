```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols = ['date', 'high', 'low', 'open', 'close']

apple_stocks = pd.read_csv("AAPL.csv", usecols=cols, index_col='date', parse_dates=['date'])
apple_stocks
```

Plot some specific values from dataframe

```python
# we need to use .loc method
# as we already made date column as index, so lets see first how that dataframe looks like
apple_stocks.head()
```

```python
# so now we can select values using index
apple_stocks.loc['2016-05'].plot()
```

now even above is data for 2016 but included all the columns.
To select specific columns 

```python
apple_stocks.plot(
    y = 'close' # defining y-axis and this can be a list of columns or a single column
    )
```

```python
apple_stocks.plot(
    y = ['close', 'open'] # defining y-axis and this can be a list of columns or a single column
    )
```

```python
# want plot on 2016 year of data
apple_stocks.loc['2016-05'].plot(
    y = ['close', 'open'] # defining y-axis and this can be a list of columns or a single column
    )
```

```python
# Getting one plot for each column
# for this we just need to set subplots to True

apple_stocks.plot(subplots=True)
```

```python
# we can resize it as well with - figsize parameter
apple_stocks.plot(subplots=True, figsize=(20, 20)) # figsize takes dimensions in tuples
```

Modifying Plot informations like 
1. Setting Title

```python
apple_stocks.plot()
plt.title("Apple stocks")
```

```python
# setting X label and Y label
apple_stocks.plot()
plt.title("Apple stocks")
plt.xlabel("Years")
plt.ylabel("Price")
```

```python
# getting grid in plot
apple_stocks.plot()
plt.title("Apple stocks")
plt.xlabel("Years")
plt.ylabel("Price")
plt.grid()
```

Above we noticed, column names are given in color - 
1. Close - is in blue
2. high - is in orange
3. low - is in green
4. open - is in red

but we can also change this color with `c` parameter

```python
# want close column in green color and data should be picked from 2016
apple_stocks.loc['2016-05'].plot(
    y='close',
    c='g'
)
```

```python
# we can increase the width of the line as well
apple_stocks.loc['2016-05'].plot(
    y='close',
    c='g',
    linewidth=5
)
```

```python
# we can change the style of line as well
apple_stocks.loc['2016-05'].plot(
    y='close',
    c='g',
    linewidth=5,
    linestyle=":" # want in dotted
)
```
