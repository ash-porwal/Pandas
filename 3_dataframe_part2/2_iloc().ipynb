### iloc[]

So, what is difference between iloc and loc?

`loc[]` allows us to use labels(rows) to select rows
* you can use label names, and if the index is numerical, you can use numbers

`iloc[]` allows us to use index position to select rows
* you can use numbers to access elements via there index position

```python
import pandas as pd
```

```python
df = pd.read_excel('titanic3.xls')
df.head()
```

```python
# selecting second row
df.iloc[1]
```

```python
# we can slice them too
df.iloc[0:3]
```

```python
# lets try to get the same with loc
df.loc[1]
```

### So, what is difference between iloc and loc?
as we saw output were the same
Well here is the thing with loc[index] we need to specify the name of the index
So, if I had taken names column as index then I would have to use that name to get that row data.
but in case of iloc we still could use numbers to get the row

```python
# lets see
df = pd.read_excel('titanic3.xls', index_col='name')
df.head(3)
```

```python
# now lets try to get the second tow with iloc
df.iloc[3]
```

```python
# now lets try with loc
df.loc[3] 
# see? 
# it is a KeyError - so to make it work we had to define it with index name
```

```python
# getting 3rd low with loc
df.loc['Allison, Miss. Helen Loraine']
```

```python
# we can also get the specific multiple rows
df.iloc[[121,124,50]]
```

```python
# to know last 3 rows
df.iloc[-3:]
```

```python
# now lets try to get 9th value from 3th column
df.iloc[9, 3] # [index postion, column position] # if it returns nothing then value is 0
```

```python
# if we want to get multiple columns and indexes value then we can do this
df.iloc[[2,3,4], [6,7,8]]
```
