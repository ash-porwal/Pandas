### Renaming

```python
import pandas as pd
```

```python
df = pd.read_excel('titanic3.xls')

# lets list out all the columns
list(df.columns)
```

Renaming columns

```python
# lets fist get the column which we want to rename
df.columns[0]
```

```python
# lets try to rename pclass to class
df.columns[0] = 'Class' # but this will raise exception
```

One way to rename columns is to pass in a list of names of the same length

```python
df.columns = ['a', 'b', 'c', 'd', 'e', 'sibsp','parch','ticket','fare','cabin','embarked','boat','body', 'home.dest']
df.head(2)
```

Better way is to use rename()

```python
df.rename(columns={
    # orginal column name: #the column with which we want to replace
    'Ticket price': 'Fare'
    }, inplace=True)
df
```

So, by default all the operations are performed on the index(axis is 0).
and we can choose axis based on which we want to perform this change

```python
df.rename({'fare': 'Cost', 'Class': 'Groups'}, axis=1)
```
