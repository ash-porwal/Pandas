### loc()
* With this method we can select the rows in a dataframe

```python
import pandas as pd
```

```python
df = pd.read_excel('titanic3.xls')
df.head()
```

```python
# but I want to read the name column as index
df = pd.read_excel('titanic3.xls', index_col='name')
df
```

```python
# as we are selecting the rows,
# so when index are sorted then selecting rows is faster.
df.sort_index(inplace=True)
df
```

```python
# now we can get the row by index
df.loc['Abbing, Mr. Anthony'] # this returns a series
```

```python
# if we want to get the multiple rows then
# we could pass a list of indexes name
df.loc[['Abbing, Mr. Anthony', 'van Billiard, Mr. Austin Blyler']] # and muliple rows returns df
```

```python
# what happens if we try to get that row which doesn't exists
df.loc['wont_exist'] # gonna throw KeyError
```

```python
# in my file I've two persons got the same name
# so if we use loc method on their name then what would happen
df.loc['Kelly, Mr. James']
# so it is going to return all the rows with that name
```

```python
# so, suppose if we just want the know fare cost for above person
# then we can do this loc too
df.loc['Kelly, Mr. James', 'fare']
```

```python
# if I wanted to know about multiple columns
# for that we just need to pass a list of columns with the index(in my case index is name)
df.loc['Kelly, Mr. James', ['fare', 'ticket']]
```

```python
# if I want to know multiple columns info for multiple passengers
# I would pass a list of index and a list of column
df.loc[['Kelly, Mr. James', 'Abbing, Mr. Anthony'], ['fare', 'ticket']]
```
