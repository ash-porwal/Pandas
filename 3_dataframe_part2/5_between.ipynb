```python
import pandas as pd
```

```python
cols = ['name', 'survived', 'pclass', 'fare', 'age', 'sex']
```

```python
df = pd.read_excel('titanic3.xls', usecols=cols, index_col='name')
df.head()
```

So, if i had filter those people who are more than 20 and less than 25 then I would do like below

```python
age_condition = df['age'] >= 20
age_cond_upper = df['age'] <= 25

df[age_condition & age_cond_upper]
```

Another way to do above is with -> `between`

```python
youth_mask = df['age'].between(20,25)
df[youth_mask]
```

You also have an option to define whether you wnt to include `both` ends, or only keep what is in between using `neither`

```python
youth_mask = df['age'].between(20,25, inclusive='neither') # by default inclusive is set to both
df[youth_mask]
```
