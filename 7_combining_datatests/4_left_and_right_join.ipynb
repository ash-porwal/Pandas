```python
import pandas as pd
```

```python
cols = ["Name", "Salary"]
company21 = pd.read_csv(
    "company21.csv", 
    sep=';',
    usecols=cols
    )
company22 = pd.read_csv(
    "company22.csv", 
    sep=';',
    usecols=cols
    )
```

Left join - **right side df** will join on **left side df**, means if value is not present in **right side df** then it will show up as `NaN` in the output, but it will have all **left side df** values

```python
pd.merge(
    left=company21,
    right=company22,
    how='left',
    on='Name',
    suffixes=("_21", "_22"),
    indicator=True
)
```

Right join - reverse of Left join

```python
pd.merge(
    left=company21,
    right=company22,
    how='right',
    on='Name',
    suffixes=("_21", "_22"),
    indicator=True
)
```
