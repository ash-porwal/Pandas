```python
import pandas as pd
```

```python
cols = ["Name", "Salary"]
company21 = pd.read_csv(
    "company21.csv", 
    sep=';',
    usecols=cols
    )
company22 = pd.read_csv(
    "company22.csv", 
    sep=';',
    usecols=cols
    )
```

Inner join - will contain those values which both dataframes will have

```python
pd.merge(
    left=company21,
    right=company22,
    how='inner', # by default it is inner
    on="Name",  # Column name on which we want to perform operation
    suffixes=('_21', '_22'),
    indicator=True
)
```

if we perform operation on Salary column, then it will look for people who has the exact same salary in both the dataframes.

```python
pd.merge(
    left=company21,
    right=company22,
    how='inner', # by default it is inner
    on="Salary",  # Column name on which we want to perform operation
    suffixes=('_21', '_22'),
    indicator=True
)
```

but if we dont specify any column name to perform join, then it is by **default** going to be `None` and it is going to show those values of both columns which are exact same(as our dataframe contains `Salary` and `Name` column).

Means - **will show that result from df1 and df2 where Salary and Name value is exact same**

```python
pd.merge(
    left=company21,
    right=company22,
    how='inner', # by default it is inner
    suffixes=('_21', '_22'),
    indicator=True
)
```
