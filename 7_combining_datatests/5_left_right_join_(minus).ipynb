```python
import pandas as pd
```

```python
cols = ["Name", "Salary"]
company21 = pd.read_csv(
    "company21.csv", 
    sep=';',
    usecols=cols
    )
company22 = pd.read_csv(
    "company22.csv", 
    sep=';',
    usecols=cols
    )
```

Just performing Outer merge to see output

```python
pd.merge(
    left=company21,
    right=company22,
    how='outer', 
    on="Name", 
    suffixes=('_21', '_22'),
    indicator=True 
)
```

Notice - on above dataset we have merge info
1. 3 left_only
2. 4 right_only
3. and rest are both 

Now, to perform a **left (minus)** or **right (minus)** join - we need to create `mask`

And we want **left (minus)** when we just want to see left side only dataset values and should not be present in both of the datasets

```python
my_merged_df = pd.merge(
    left=company21,
    right=company22,
    how='outer', 
    on="Name", 
    suffixes=('_21', '_22'),
    indicator=True 
    )

# extracting values on _merge and making a condition
my_merged_df[my_merged_df['_merge'] == 'left_only']
```

and for `right_only` we will just make condition equal to `right_only`

```python
my_merged_df[
    my_merged_df['_merge'] == 'right_only'
    ]
```
