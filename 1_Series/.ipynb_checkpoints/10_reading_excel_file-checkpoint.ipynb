### Reading Excel file

So, there are two types of excel file,
1. New formatted
2. Old formatted

* New excel files they end with extensin - .xlsx
* old excel file they end with - .xls

```python
import pandas as pd
```

```python
# Reading new excel files, we can read new file with pandas only
df = pd.read_excel('file_example_XLSX_10.xlsx')
df
```

```python
# Now, to read old excel which ends with xls, they will need
# xlrd install
# openypyxl (it is pre installed in conda environment)
# after installing above we can read it as usual
df_old_excel = pd.read_excel('file_example_XLS_10.xls')
df_old_excel
```
