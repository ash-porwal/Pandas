```python
import pandas as pd
```

```python
df = pd.read_csv("Pokemon.csv", index_col='#')
df.head()
```

So, far we have learnt that we can apply string operations on dataframe using .str
and we cannot use those operations directly because they were considered as objects.

Again - Strings are objects but objects are not strings.
Now, let's explore a few more methods that we can use with strings.

.startswith()

```python
name_filter = df['Name'].str.startswith('Do') # remember? it is case sensitive
df[name_filter]
```

.endswith()

```python
name_filter = df['Name'].str.endswith('chu') # remember? it is case sensitive
df[name_filter]
```

.replace()

```python
df['Type 1'].str.replace("Grass", "Vegan", regex=False)
```
