```python
import pandas as pd
```

```python
df = pd.read_csv("Pokemon.csv", index_col='#')
df.head()
```

Lets look how we can use Strings in Pandas

```python
df['Name'].upper() # Doesn't work
```

```python
df['Name'].astype(str).upper() # Doesn't work either
```

```python
# This is how it works
df['Name'].str.upper()
```

So, every time you want to use string method it is important to use .str first

```python
df['Attack'].str.upper() # Doesn't work with non string values
```

```python
df['Name'].str.lower()
```

```python
df['Name'].str[::-1] # reverse the sting
```

Can also be used for filtering

```python
x_filter = df['Name'].str.contains('q', case=False) # Whether to ignore case or not
# above we are checking if Name column contains a certain value or not.
# so we are checking if string contains q letter or not, and case=False defines that 
# it shouldn't be case sensitive.(means - q and Q both are going to be treated as same character)
df[x_filter]
```
